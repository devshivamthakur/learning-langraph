```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Optional
```

```python
class Student(TypedDict):
    name: str
    age: int
class Course(TypedDict):
    title: str
    credits: int

class main(TypedDict):
    student: Student
    course: Course
    combineResult: Optional[str]
    

```

```python
#initialize graph with state
graph = StateGraph(main)
```

```python
def add_student(state: main) -> main:
    ans = f"student {state['student']['name']} having age {state['student']['age']} is pursuing {state['course']['title']}"
    state['combineResult'] = ans
    return state
```

```python
#add nodes
graph.add_node("add_student", add_student)

#add edges
graph.add_edge(START, "add_student")

graph.add_edge("add_student", END)
```

```python
workflow = graph.compile()
result = workflow.invoke({
    'student': {'name': 'Alice', 'age': 20},
    'course': {'title': 'Math 101', 'credits': 3}
})
```
